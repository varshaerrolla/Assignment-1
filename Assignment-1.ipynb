```python
import numpy as np

# Task a1: Creating a random vector of size 15 with integers in the range 1-20
random_vector = np.random.randint(1, 21, 15)
random_vector
```

```python
# Task a2: Reshaping the array to 3 by 5
reshaped_array = random_vector.reshape(3, 5)

# Printing the array shape
reshaped_array.shape
```

```python
# Task a3: Replacing the maximum value in each row with 0
for i in range(reshaped_array.shape[0]):
    max_value_index = np.argmax(reshaped_array[i])
    reshaped_array[i][max_value_index] = 0

# Printing the array after replacing the max value with 0
reshaped_array
```

```python
# Creating a 2-dimensional array of size 4 x 3 with 4-byte integer elements
array_2d = np.empty((4, 3), dtype=np.int32)
```

```python
# Printing the shape of the array
print("2d Array shape:", array_2d.shape)

# Printing the type of the array
print("2d Array type:", type(array_2d))

# Printing the data type of the array
print("2d Array data type:", array_2d.dtype)
```

```python
# Defining the square array
square_array = np.array([[3, -2],
                         [1, 0]])

# Computing the eigenvalues and right eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(square_array)

# Printing the eigenvalues
print("Eigenvalues:")
for eigenvalue in eigenvalues:
    print(eigenvalue)

# Printing the right eigenvectors
print("\nRight Eigenvectors:")
for eigenvector in eigenvectors.T:  
    print(eigenvector)
```

```python
array = [[0, 1, 2],
         [3, 4, 5]]

# Initializing the sum variable
diagonal_sum = 0

# Computing the sum of diagonal elements
for i in range(len(array)):
    diagonal_sum += array[i][i]

# Printing the sum of diagonal elements
print("Sum of diagonal elements:", diagonal_sum)
```

```python
# Defining the original array
original_array = np.array([[1, 2],
                           [3, 4],
                           [5, 6]])

# Reshaping the array to 3x2
reshaped_array_1 = np.reshape(original_array, (3, 2))

# Reshaping the array to 2x3
reshaped_array_2 = np.reshape(original_array, (2, 3))

# Printing the reshaped arrays
print("\nReshaped Array 1 (3x2):")
print(reshaped_array_1)

print("\nReshaped Array 2 (2x3):")
print(reshaped_array_2)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Programming languages and their popularity
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

# Create a DataFrame with programming languages and their popularity
prgmng_df = pd.DataFrame({"popularity": popularity},
                         index=languages)

# Define the explode values
myexplode = [0.1, 0, 0, 0, 0, 0]

# Create the pie chart
prgmng_df.plot.pie(subplots=True, autopct='%1.1f%%', explode=myexplode, figsize=[10, 10], startangle=140, shadow=True)

# Show the pie chart
plt.show()
```
